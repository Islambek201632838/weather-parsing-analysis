```python
import pandas as pd
```

```python
# Load the data using pandas
df_pandas = pd.read_csv('weather_data.csv')
```

```python
df_pandas
```

```python
# Calculate the average temperatures
average_day_temp = df_pandas['Day Temperature'].mean()
average_night_temp = df_pandas['Night Temperature'].mean()
```

```python
# Identify the hottest and coldest days
hottest_day = df_pandas.iloc[df_pandas['Day Temperature'].idxmax()]
coldest_day = df_pandas.iloc[df_pandas['Night Temperature'].idxmin()]
```

```python
# Print the results
# Create a DataFrame for the analysis results
analysis_results = pd.DataFrame({
    "Metric": ["Average Day Temperature", "Average Night Temperature", "Hottest Day", "Coldest Day"],
    "Value": [
        f"{average_day_temp:.2f}°",
        f"{average_night_temp:.2f}°",
        f"{hottest_day['Day Temperature']}° on date {hottest_day['Date']}",
        f"{coldest_day['Night Temperature']}° on date {coldest_day['Date']}"
    ]
})

# Display the DataFrame
print("Pandas Analysis")
display(analysis_results)

```

```python
import polars as pl
```

```python
# Now using polars
df_polars = pl.read_csv('weather_data.csv')
df_polars
```

```python
# Calculating average, minimum, and maximum for day and night temperatures
avg_day_temp = df_polars['Day Temperature'].mean()
avg_night_temp = df_polars['Night Temperature'].mean()

min_day_temp = df_polars['Day Temperature'].min()
max_day_temp = df_polars['Day Temperature'].max()

min_night_temp = df_polars['Night Temperature'].min()
max_night_temp = df_polars['Night Temperature'].max()
```

```python
results = pl.DataFrame({
    "Metric": ["Average Day Temperature", "Average Night Temperature", 
               "Minimum Day Temperature", "Maximum Day Temperature",
               "Minimum Night Temperature", "Maximum Night Temperature"],
    "Value": [avg_day_temp, avg_night_temp, min_day_temp, max_day_temp, 
              min_night_temp, max_night_temp]
})

display(results)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(15, 5))  # Increase the figure size to make it wider

# Plotting the day temperatures
plt.plot(df_pandas['Date'], df_pandas['Day Temperature'], label='Day Temperature')
plt.scatter(hottest_day['Date'], hottest_day['Day Temperature'], color='red', label='Hottest Day')

# Formatting the dates on the x-axis to display less frequently
# For example, display every 7th label (one label per week if daily data)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))

# Rotating labels for better readability
plt.xticks(rotation=45)

# Setting labels and title
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Day Temperature Over Time')

# Adjustting layout to prevent label cut-off
plt.tight_layout()

# Showing the plot with adjustments
plt.legend()
plt.show()

```

```python

plt.figure(figsize=(15, 5))  # Increase the figure size to make it wider

# Plotting the day temperatures
plt.plot(df_pandas['Date'], df_pandas['Night Temperature'], label='Night Temperature')
plt.scatter(coldest_day['Date'], coldest_day['Night Temperature'], color='blue', label='Coldest Night')

# Formating the dates on the x-axis to display less frequently
# For example, displaying every 7th label (one label per week if daily data)
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))

# Rotating labels for better readability
plt.xticks(rotation=45)

# Setting labels and title
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Night Temperature Over Time')

# Adjusting layout to prevent label cut-off
plt.tight_layout()

# Showing the plot with adjustments
plt.legend()
plt.show()
```
